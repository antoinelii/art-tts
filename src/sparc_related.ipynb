```python
from model_ms.spk_encoder import SpeakerEncodingLayer
from paths import CKPT_DIR
import torch

ckpt = CKPT_DIR / "sparc_multi.ckpt"
sparc_ckpt = torch.load(ckpt)
spk_enc = SpeakerEncodingLayer(spk_ft_size=1024,spk_emb_size=64,)
spk_enc.load_state_dict(sparc_ckpt["state_dict"]["spk_ft"])
spk_enc.eval()
#spk_enc._modules["spk_fc"][0].weight
spk_ft = torch.randn(1, 1024)
with torch.no_grad():
    spk_emb = spk_enc(spk_ft)
spk_emb
```

```python
from model_ms.sparc_generator import HiFiGANGenerator, SpkHiFiGANGenerator

gen = HiFiGANGenerator(**sparc_ckpt["config"]["generator_configs"])
```

```python
#spk_enc_out_dim = sparc_ckpt["state_dict"]["spk_ft"]['spk_fc.3.weight'].shape[0]
#assert sparc_ckpt["config"]["generator_configs"]["spk_emb_size"] == spk_enc_out_dim
spkgen = SpkHiFiGANGenerator(spk_ft_size=1024,**sparc_ckpt["config"]["generator_configs"])
spkgen.spk_enc.load_state_dict(sparc_ckpt["state_dict"]["spk_ft"])
spkgen.generator.load_state_dict(sparc_ckpt["state_dict"]["generator"])
spkgen.eval()
# generate tensor of shape (batch, channels, time)
c = torch.randn(1, 14, 100)  # (batch, channels, time)
spk_ft = torch.randn(1, 1024)  # (batch, spk_ft_size)
with torch.no_grad():    
    out = spkgen(c, spk_ft)
```
